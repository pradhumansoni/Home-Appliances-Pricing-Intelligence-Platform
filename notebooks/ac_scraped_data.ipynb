```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
# Open raw ac_full_dump file

with open('D:/Study/data_science/underpriced-listing-predictor/scraper/air_conditioner.html','r' , encoding='utf-8') as f:
    ac_html = f.read()

soup = BeautifulSoup(ac_html , 'lxml') # use beautiful soup class to check ac_html file
```

```python
containers = soup.find_all('div' , {'class':'sm-product has-tag has-features has-actions'}) # Containers containing each card data 

print(f"Total Cards -> {len(containers)}")
```

```python
# We see that The names of each appliance is inside "h2"
name = []
price = []
rating = [] 
ac_type = []
capacity = []
features = []



for i in containers:
    try:
        name.append(i.find("h2").text)
    except:
        name.append(np.nan)
    try:
        price.append(i.find('span' ,{'class':'price'} ).text)
    except:
        price.append(np.nan)
    try:
        rating.append(i.find('div', {'class': 'rating'}).find('span' ,{'class':'sm-rating'})['style'])
    except:
        rating.append(np.nan)
    x = i.find('ul' , {'class': 'sm-feat specs'}).find_all('li')
    try:
        ac_type.append(x[0].text)
    except:
        ac_type.append(np.nan)
    try:
        capacity.append(x[1].text)
    except:
        capacity.append(np.nan)

    sub_feature = []
    
    try:
        sub_feature.append(x[2].text)
    except:
        sub_feature.append(np.nan)
    try:
        sub_feature.append(x[3].text)
    except:
        sub_feature.append(np.nan)
    try:
        sub_feature.append(x[4].text)
    except:
        sub_feature.append(np.nan)
    try:
        sub_feature.append(x[5].text)
    except:
        sub_feature.append(np.nan)
    try:
        sub_feature.append(x[6].text)
    except:
        sub_feature.append(np.nan)
    try:
        sub_feature.append(x[7].text)
    except:
        sub_feature.append(np.nan)
    features.append(sub_feature)
    
```

```python
import pandas as pd

df = pd.DataFrame({'name':name , 'price':price, 'rating': rating , 'ac_type':ac_type , 'capacity':capacity , 'features':features})

print(df.info())
```
